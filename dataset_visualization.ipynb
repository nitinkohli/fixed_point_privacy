## Generation of Figure 1

```python
fix,ax = plt.subplots(1,2,figsize= (5,1.3))
plt.subplots_adjust(wspace=.5)

count_table = [['State','Cases'],
               ['Alabama','0'],
               ['...','...'],
               ['California','38'],
               ['Colorado','10'],
               ['...','...'],
               ['Wisconsin','1'],
               ['Wyoming','0']
              ]

ax[0].set_axis_off()
table = ax[0].table(cellText= count_table, loc='upper center')

## Set the linewidth of all cells
for (row, col), cell in table.get_celld().items():
    cell.set_linewidth(.5)  

table._cells[(0, 0)].set_facecolor("lightgray")
table._cells[(0, 1)].set_facecolor("lightgray")
ax[0].set_title('A: Table of Counts', fontsize = 10, pad = 8)

ax[1].hist(bird_flu.Bird_Flu_Cases, bins = 39, density =True)
ax[1].set_xticks(range(0,41,10))
ax[1].set_xticklabels([str(i) for i in range(0,40,10)] + ['40'])
ax[1].set_xlabel("Count of Cases",fontsize=9)
ax[1].set_ylabel("Fraction of States", fontsize=9, labelpad=0)
ax[1].set_title('B: Distribution of Counts', fontsize = 10, pad = 8, loc = "right")


plt.show()
```
